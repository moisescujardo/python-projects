```python
# import pandas and give it an alias (pd)
import pandas as pd

# next, let's import pyplot from matplotlib
# give it an alias of plt
from matplotlib import pyplot as plt
```

```python
# read intel-stock-data.csv file
# read_csv will return a DataFrame object
intel_stock_df = pd.read_csv('data/psei-data.csv')

# let's validate the type of intel_stock_df
type(intel_stock_df)
```

```python
# a DataFrame is a row and column of data
intel_stock_df
```

```python
# let's take a look at the shape 
# of our DataFrame
# shape is a property shows the 
# number of rows and columns (r, c)
# in the dataframe
intel_stock_df.shape
```

```python
# the columns attribute and array
# of column names with in the dataframe
type(intel_stock_df.columns)
```

```python
# let's get some information 
# about our DataFrame
intel_stock_df.info()

# the first line tells us that it's a DataFrame
# There are 254 rows. Range Index is 0 to 253
# it shows the columns, non-null count 
# and the datatype of each columns
```

```python
# the describe function gives you 
# some statistical information about
# about the DataFrame.
intel_stock_df.describe()
```

```python
# you can also take a look at the first nRows
# with in the DataFrame by using the head(n) function
# here we're loading the first 10 rows from the DataFrame
intel_stock_df.head(10)
```

```python
# you can also take a look at the bottom nRows
# in the DataFrame. You can do this using the tail(n)
# function
intel_stock_df.tail(10)
```

```python
# A Series is a column in the DataFrame
# you can access a series by using DataFrame['column_name']
# example:
intel_stock_df['Open']

# note: column name is case sensitive meaning,
# it has to be thesame with the column name 
# on the CSV file.
# intel_stock_df['open'] will generate an error:
# KeyError: 'open'
```

```python
# here's another example where I extract 3 columns
# from the DataFrame and get only the first 20 rows
intel_stock_df[['Date', 'Open', 'High', 'Low']].head(20)
```

# Plotting Basics

```python
# this is the default plot
intel_stock_df.plot()
```

```python
# let's try to plot this intel stock data
# plot line graph from our DataFrame
intel_stock_df.plot.line()
```

```python
# setting subplots=True will plot
# each axis individually
intel_stock_df.plot.line(subplots=True)
```

```python
# each chart or graph has its own default colors
# that can be overriden by specifying color values (dict)


# remove the subplots=True and 
# you will get one graph with different
# colors from the default

# let's continue where we left
axis = intel_stock_df.plot.line(
    subplots=True,
    x='Date',
    color={
        'Open': '#a4cb47',
        'High': '#f3bb69',
        'Low': '#599849',
        'Close': '#aed0a7',
        'Adj Close': '#000000',
        'Volume': '#ff3344'
    }
)
```

```python
# here's a line chart with close values by date
intel_stock_df.plot(
    kind='line', x='Date', y=['Open', 'Close'],
    color={
        'Open': '#a4cb47',
        'Close': '#599849'
      }
)
```

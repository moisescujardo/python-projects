```python
# you can also take a look at the first nRows
# with in the DataFrame by using the head(n) function
# here we're loading the first 10 rows from the DataFrame
psei_df.head(10)
```

# Plotting Basics

```python
# this is the default plot
psei_df.plot()
```

```python
# let's try to plot this intel stock data
# plot line graph from our DataFrame
psei_df.plot.line()
```

```python
# setting subplots=True will plot
# each axis individually
psei_df.plot.line(subplots=True)
```

```python
# each chart or graph has its own default colors
# that can be overriden by specifying color values (dict)


# remove the subplots=True and 
# you will get one graph with different
# colors from the default

# let's continue where we left
axis = psei_df.plot.line(
    subplots=True,
    color={
        'Open': '#a4cb47',
        'High': '#f3bb69',
        'Low': '#599849',
        'Close': '#aed0a7',
        'Adj Close': '#000000',
        'Volume': '#ff3344'
    }
)
```

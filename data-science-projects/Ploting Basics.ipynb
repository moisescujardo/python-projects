```python
# import pandas
import pandas as pd
# import pyplot
import matplotlib.pyplot as plt
```

```python
# open csv file (to DataFrame)
# index_col = 0 indicate the first column as index
# parse_dates = True convert dates
air_quality_data = pd.read_csv('data/air_quality_no2.csv', index_col=0, parse_dates=True)
```

```python
# display default plot
air_quality_data.plot()
```

```python
# plot by column
# on this case, plot the data in station_paris
air_quality_data['station_paris'].plot()
```

```python
# how about station_london? Let's plot the data
air_quality_data['station_london'].plot()
```

```python
# let's visually compare the No2 values between station_londo and station_paris
air_quality_data.plot.scatter(x = 'station_london', y = 'station_paris', alpha=0.3)
```

```python
# let's try to use the boxplot
air_quality_data.plot.box()
```

```python
axs = air_quality_data.plot.area(figsize=(12, 4), subplots=True)
plt.show()
```

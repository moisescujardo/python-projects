# Filtering (Conditional Filtering) DataFrame

```python
# let's try to perform some filtering
# but first, let's check what's inside our DataFrame
intel_stock_df
```

```python
# Open that are greater than 40
# this will return an array of boolean
open_greater_than_40 = intel_stock_df['Open'] > 40
# probably because, Pandas reads the dataframe and
# return if the condition is true or not
open_greater_than_40.dtype
```

```python
# to show the data, we can do the line below
intel_stock_df[open_greater_than_40]

# notice that the DataFrame now contains
# rows whos open price is greater than 40
```

```python
# Here's another approach that will return the same result
open_greater_than_40 = intel_stock_df[intel_stock_df['Open']>40]
open_greater_than_40
```

```python
# let's try another example
high_less_than_40 = intel_stock_df[intel_stock_df['High'] < 40]
high_less_than_40
```

```python
# here, we're filtering using two column arguments
open_gt_39_close_lt_40 = intel_stock_df[(intel_stock_df['Open']>30) & ((intel_stock_df['Close']<=40))]
open_gt_39_close_lt_40
```

# Plotting Basics

```python
# this is the default plot
intel_stock_df.plot()
```

```python
# let's try to plot this intel stock data
# let's try using the line chart first.

# plot line graph from our DataFrame
intel_stock_df.plot.line()
```

```python
# setting subplots=True will plot
# each axis individually
axis = intel_stock_df.plot.line(subplots=True)
# this will return numpy.ndarray
type(axis)
```

```python
# each chart or graph has its own default colors
# that can be overriden by specifying color values (dict)


# remove the subplots=True and 
# you will get one graph with different
# colors from the default

# let's continue where we left
axis = intel_stock_df.plot.line(
    subplots=True,
    color={
        'Open': '#a4cb47',
        'High': '#f3bb69',
        'Low': '#599849',
        'Close': '#aed0a7'
        
    }
)
```

```python
# here's a line chart with close values by date
intel_stock_df.plot(kind='line', x='Date', y='Open')
```

```python
# you can also specify multiple series on the Y axis
# for example, on the Y axis, I want to show the 
# Open, Close, High and Low stock values
# but, it's using the default color scheme
intel_stock_df.plot(kind='line', x='Date', y=['Open', 'Close', 'High', 'Low'])

# Note: Some graphs - like kde, density - will require scipy module
```

```python
# let's experiment and place in our own color scheme
# same line from above but custom color scheme
intel_stock_df.plot(
    kind='line', x='Date', y=['Open', 'Close', 'High', 'Low'],
    color={
        'Open': '#a4cb47',
        'High': '#f3bb69',
        'Low': '#599849',
        'Close': '#aed0a7'
        
    }
)
```
